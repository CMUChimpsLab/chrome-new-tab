```python
# imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# user defined classes
import DataLoader

dataLoader = DataLoader.DataLoader()

# read data file name
filename = dataLoader.read_filename()
print('Using data file: %s' % filename)

raw_data = dataLoader.read_df()
# print(raw_data)
```

## Chapter 1: User-level analysis

In this section, we consider user attributes as independent variables.

### Number of matches
A *match* happens when the setting option selected is the same as the user's current setting on Facebook. This measure applies only to conditions 2 and 3, because we did not ask for expected values in condition 1.

```python
depv = 'num_matches'

# create box plot
raw_data.boxplot(depv, by='condition', figsize=(12, 8))

# get data per condition
c1, c2, c3 = dataLoader.depv_per_condition(depv)

# calculate the t-test for the means of two independent samples
t, p = stats.ttest_ind(c2, c3)
print('t = %f, p = %f' % (t, p))

# run one-way ANOVA and display results
F, p = stats.f_oneway(c2, c3)
print('F = %f, p = %f' % (F, p))
```

### Scenario confidence
For this measure, we asked users to rate their confidence level on the following scenarios:
 
* **Scenario #1: Bob**  
Bob, who is a close friend of yours, recently received a friend request from a stranger. Without Bob accepting the request, this stranger posted highly inappropriate content on his timeline. All his friends were able to see and comment on the posts.  Because of this incident, Bob was fired from his job.  

* **Scenario #2: Craig**  
Craig went to party last weekend and drank too much. His friend took a video of him talking bad things about the company where he works and posted on Facebook. Facebook automatically recognized Craig on the video and added it to his timeline. On the following day, his boss saw this video.  

* **Scenario #3: Anna**  
Your friend Anna complained to you about receiving 20 spam phone calls in a day. People from the advertising companies claimed that they found her number on Facebook by looking up her name on a search engine.  

How confident are you that your Facebook account is protected against these scenarios?  
1 = Not confident at all  
2 = Mostly not confident  
3 = Neutral  
4 = Mostly Confident  
5 = Strongly Confident


```python
depv = 'scenarios_confidence'

# create box plot
raw_data.boxplot(depv, by='condition', figsize=(12, 8))

# get data per condition
c1, c2, c3 = dataLoader.depv_per_condition(depv)

# run one-way ANOVA and display results
F, p = stats.f_oneway(c1, c2, c3)
print('F = %f, p = %f' % (F, p))
```
